```python
import os 
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
fire_path = "../data/fire"
no_fire_path = "../data/no_fire"

fire_count = len(os.listdir(fire_path))
no_fire_count = len(os.listdir(no_fire_path))

print(f"Fire images:    {fire_count}")
print(f"No fire images: {no_fire_count}")
print(f"Total:          {fire_count + no_fire_count}")
```

```python
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle("Sample images from dataset", fontsize=14)

# top row = fire
fire_imgs= random.sample(os.listdir(fire_path),5)
for i, fname in enumerate(fire_imgs):
    img=Image.open(os.path.join(fire_path,fname))
    axes[0, i].imshow(img)
    axes[0, i].set_title("fire", color="red")
    axes[0, i].axis("off")
# bottom row = no fire
no_fire_imgs = random.sample(os.listdir(no_fire_path), 5)
for i, fname in enumerate(no_fire_imgs):
    img = Image.open(os.path.join(no_fire_path, fname))
    axes[1, i].imshow(img)
    axes[1, i].set_title("no fire", color="green")
    axes[1, i].axis("off")

plt.tight_layout()
plt.show()
```

```python
sizes = []
for fname in os.listdir(fire_path)[:50]:
    img = Image.open(os.path.join(fire_path, fname))
    sizes.append(img.size)

unique_sizes = set(sizes)
print(f"Unique sizes in fire folder (first 50): {unique_sizes}")
print("(All will be resized to 224x224 during training — this is just a check)")
```
